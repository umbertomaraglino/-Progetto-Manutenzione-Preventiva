```python
#UNITO DATAFRAME CASE
import os
import pandas as pd
folder_path = r"C:\Users\umber\Desktop\data"  # Assumendo che i file siano nella cartella corrente, puoi modificare il percorso se necessario
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

# Inizializza una lista per i DataFrame
dataframes = []

# Loop attraverso ogni file CSV
for csv_file in csv_files:
   
    
    # Caricare il file CSV in un DataFrame Pandas
    df = pd.read_csv(os.path.join(folder_path, csv_file))
    
    # Aggiungi il DataFrame alla lista
    dataframes.append(df)

# Unisci tutti i DataFrame in uno solo
merged_df = pd.concat(dataframes, ignore_index=True)

# Salva il DataFrame unificato in un nuovo file CSV
merged_df.to_csv(r"C:\Users\umber\Desktop\merged_data.csv", index=False)
```
