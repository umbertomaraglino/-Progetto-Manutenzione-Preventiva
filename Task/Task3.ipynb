```python
# diversi modelli
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
x_test=pd.read_csv(r"C:\Users\umber\Desktop\MANUTENZIONE\dataset_test_finale.csv")
x_test= x_test[["Case","P1","P2","P3","P4","P5","P6","P7"]]
desired_values = [178, 186, 193, 196, 197, 204, 209, 219, 221]
x_test = x_test[x_test['Case'].isin(desired_values)]
test = x_test[["P1","P2","P3","P4","P5","P6","P7"]]
colonna = x_test[["Case"]]
train_df = pd.read_csv(r"C:\Users\umber\Desktop\MANUTENZIONE\dataset_train_finale.csv")
train_df = train_df[["Condition","BP1","BP2","BP3","BP4","BP5","BP6","BP7","BV1","P1","P2","P3","P4","P5","P6","P7"]]
train_df = train_df[train_df["Condition"]== 'Anomaly'] 

colonne_da_modificare = ["BP1", "BP2", "BP3", "BP4", "BP5", "BP6", "BP7", "BV1"]
train_df[colonne_da_modificare] = train_df[colonne_da_modificare].replace({"Yes": 0, "No": 1}).astype(int)
train = train_df[["BP1", "BP2", "BP3", "BP4", "BP5", "BP6", "BP7", "BV1"]]
colonne = train_df[["P1","P2","P3","P4","P5","P6","P7"]]
train["PosizioneErrore"] = (train == 0).apply(lambda x: x.argmax() +1 if 1 in x.values else len(x), axis=1)
df_mom = pd.concat([train, colonne], axis=1)
train_df = df_mom[["PosizioneErrore","P1","P2","P3","P4","P5","P6","P7"]]
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
x = train_df[["P1","P2","P3","P4","P5","P6","P7"]]
y = train_df[["PosizioneErrore"]]
# Divisione dei dati in set di training e test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


# Definizione dei modelli
models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('AdaBoost', AdaBoostClassifier()),
    ('Gradient Boosting', GradientBoostingClassifier()),
    ('Linear SVM', LinearSVC()),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('Gaussian Naive Bayes', GaussianNB()),
    ('Linear Discriminant Analysis', LinearDiscriminantAnalysis()),
    ('Quadratic Discriminant Analysis', QuadraticDiscriminantAnalysis()),
    ('Multi-layer Perceptron', MLPClassifier())
]

# Addestramento e valutazione dei modelli
for model_name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{model_name}: Accuracy = {accuracy:.2f}')
```
